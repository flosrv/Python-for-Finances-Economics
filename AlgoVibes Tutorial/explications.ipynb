```python
"""
Open Prix d’ouverture (USD)  
   - Explication : Prix auquel l’action s’est échangée lors de l’ouverture du marché.  
   - Utilité : Sert à analyser les variations entre le début et la fin de la journée de trading.  

High Prix le plus haut (USD)  
   - Explication : Prix le plus élevé de l’action durant la journée de trading.  
   - Utilité : Indicateur de la volatilité et de la force de la tendance.  

Low Prix le plus bas (USD)  
   - Explication : Prix le plus bas atteint durant la journée.  
   - Utilité : Indicateur de la volatilité et du potentiel de support.  

Close Prix de clôture (USD)  
   - Explication : Prix final de l’action au moment de la clôture du marché.  
   - Utilité : Le prix de clôture est le plus utilisé 
   pour les graphiques et indicateurs techniques.  

Volume Nombre d’actions échangées (actions)  
   - Explication : Nombre d'actions qui ont été échangées durant la journée.  
   - Utilité : Mesure de l'intérêt du marché et de la liquidité. 
   
   Un volume élevé peut signaler un changement de tendance.  

Dividends Montant du dividende (USD)  
   - Explication : Montant distribué aux actionnaires sous forme de dividendes.  
   - Utilité : Source de revenus passifs pour les investisseurs. Sert à évaluer la rentabilité d'un investissement.  

Stock Splits Fractionnement d’actions (n° d’actions après fractionnement)  
   - Explication : Nombre d’actions après un fractionnement (split) d'actions.  
   - Utilité : Aide à comprendre la dilution des actions, mais la valeur totale détenue par un investisseur ne change pas.  

Price Prix actuel (USD)  
   - Explication : Le prix de l’action à l'instant T.  
   - Utilité : Sert de référence pour l'achat ou la vente immédiate.  

Target Price Prix cible des analystes (USD)  
   - Explication : Le prix estimé par les analystes pour l’action.  
   - Utilité : Sert à évaluer si l’action est sous-évaluée ou surévaluée 
   par rapport au marché.  

P/E Ratio prix/bénéfice (sans unité)  
   - Explication : Indicateur de combien un investisseur paie pour chaque dollar 
   de bénéfice de l'entreprise.  
   - Utilité : Permet d'évaluer la valorisation de l'entreprise. 
   Un P/E trop élevé peut signaler une surévaluation.  

Forward P/E Ratio prix/bénéfice futur (sans unité)  
   - Explication : Estimation du ratio P/E basé sur les bénéfices futurs attendus.  
   - Utilité : Permet d’évaluer les attentes de croissance future. Utilisé pour comparer à la valorisation actuelle (P/E).  

P/B Ratio prix/valeur comptable (sans unité)  
   - Explication : Comparaison entre le prix de l’action et sa valeur comptable.  
   - Utilité : Si < 1, cela peut signaler une sous-évaluation.  

ROE Rentabilité des fonds propres (%)  
   - Explication : Mesure de la rentabilité de l'entreprise 
   par rapport aux fonds propres des actionnaires.  

   - Utilité : Indicateur de la performance financière et de l'efficacité de la gestion.  

ROA Rentabilité des actifs (%)  
   - Explication : Mesure de la rentabilité des actifs totaux de l’entreprise.  
   - Utilité : Permet d’évaluer l’efficacité opérationnelle de l’entreprise.  

Debt/Equity Ratio d’endettement (sans unité)  
   - Explication : Comparaison de la dette de l’entreprise par rapport aux fonds propres.  
   - Utilité : Indicateur du risque financier. Un ratio élevé indique un risque accru lié à l’endettement.  

Margins Marge bénéficiaire nette (%)  
   - Explication : Indicateur de la proportion du revenu qui reste après les dépenses.  
   - Utilité : Plus la marge est élevée, plus l'entreprise est efficace dans la gestion de ses coûts.  

Revenue Growth Croissance du chiffre d’affaires (%)  
   - Explication : Croissance en pourcentage des ventes d'une année sur l'autre.  
   - Utilité : Mesure la capacité d’une entreprise à augmenter ses revenus. Critique pour les entreprises en croissance.  

Earnings Growth Croissance des bénéfices (%)  
   - Explication : Croissance en pourcentage des bénéfices nets d'une année sur l'autre.  
   - Utilité : Indicateur de la rentabilité future et du potentiel d'augmentation des dividendes.  

Dividend Yield Rendement du dividende (%)  
   - Explication : Le rendement annuel en dividendes par rapport au prix actuel de l'action.  
   - Utilité : Attirer les investisseurs recherchant des revenus passifs.  

Free Cash Flow Flux de trésorerie libre (USD)  
   - Explication : Trésorerie disponible après avoir couvert les dépenses d’investissement.  
   - Utilité : Montre la santé financière et la capacité à réinvestir ou à distribuer des dividendes.  

Market Cap Capitalisation boursière (USD)  
   - Explication : La valeur totale des actions en circulation d’une entreprise.  
   - Utilité : Sert à classer les entreprises (small cap, mid cap, large cap) et à estimer la taille relative d'une entreprise.  

Sector Secteur d’activité (texte)  
   - Explication : Le secteur économique dans lequel l’entreprise opère (ex : Technologie, Santé).  
   - Utilité : Permet de comparer les entreprises d’un même secteur.  

Industry Industrie spécifique (texte)  
   - Explication : Branche spécifique de l'industrie dans laquelle l’entreprise évolue (ex : semi-conducteurs, banques).  
   - Utilité : Comparaison plus détaillée entre entreprises d'un même domaine plus précis.  
"""
```
